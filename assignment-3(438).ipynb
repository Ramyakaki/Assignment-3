```python
import pandas as pd
data = pd.read_csv("africa_recession.csv")
data
```

```python
data.describe()
```

```python
q1=data['ccon'].quantile(0.25)
q3=data['ccon'].quantile(0.75)
iqr=q3-q1
print(iqr)
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
print(lower_bound)
print(upper_bound)
```

```python
outliers=data[(data['ccon']>lower_bound)|(data['ccon']<upper_bound)]
print(outliers)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(data['ccon'])
```

```python
plt.boxplot(outliers['ccon'])
```

```python
o=data[(data['ccon']<lower_bound)|(data['ccon']>upper_bound)]
print(o)
```

```python
plt.boxplot(o['ccon'])
```

An online dataset was collected from kaggle that was loaded to pandas dataframe and using head(5) displayed first 5 rows.Missing values were identified and they were filled with mean() of the column by using fillna() method.
Identified the duplicate values from the data using duplicated() method and only the unique values were printed by using drop_duplicates() method.
The process of cleaning data impact the accuracy and reliability of the insights derived from data to remove the duplicates, to filter unmanted outliers and to handle missing data.

Outlier detection : It is the process of detecting outliers, or a data point that is far away from the average, and depending on what you are trying to accomplish, potentially removing or resolving them from the analysis to prevent any potential skewing. Outlier detection is one of the most important processes taken to create accurate and reliable data.
